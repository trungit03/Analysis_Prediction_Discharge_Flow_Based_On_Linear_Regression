```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
import pandas as pd

df =pd.read_csv("F:/do_an/dacn_nhom12/data/data.csv")
df
```

```python
targets = df[['Qxt', 'Qxm', 'Tqx']]
targets
```

```python
features = df.drop((['tên hồ', 'Tqx', 'Qxm', 'Qxm']), axis=1) 
# 'station' là cột tên hồ thủy điện
s = features.columns
# Chuyển đổi cột tên hồ thủy điện thành các đặc trưng nhị phân
encoder = OneHotEncoder(sparse_output=False, drop='first')  # 'drop' để tránh đa cộng tuyến
station_encoded = encoder.fit_transform(df[['tên hồ']])

# Kết hợp các đặc trưng đã mã hóa với các đặc trưng số khác
features_encoded = pd.concat([df[s], pd.DataFrame(station_encoded, columns=encoder.get_feature_names_out(['tên hồ']))], axis=1)

# Tiền xử lý dữ liệu
X_train, X_test, y_train, y_test = train_test_split(features_encoded, targets, test_size=0.2, random_state=42)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
y_test
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, mean_absolute_error
import numpy as np
y_train_qxt = y_train['Qxt']
y_test_qxt = y_test['Qxt']
model_Qxt = LinearRegression()
model_Qxt.fit(X_train, y_train_qxt)
print(model_Qxt.score(X_test, y_test_qxt))

# Dự đoán và đánh giá mô hình
y_pred_qxt = model_Qxt.predict(X_test)
y_pred = model_Qxt.predict(X_test)
print('MSE:', mean_squared_error(y_test_qxt, y_pred_qxt))
print('RMSE:', np.sqrt(mean_squared_error(y_test_qxt, y_pred_qxt)))
print('MAE:', mean_absolute_error(y_test_qxt, y_pred_qxt))
print(f'MAPE: {mean_absolute_percentage_error(y_test_qxt, y_pred_qxt)*100}%')
results = pd.DataFrame({'y_test': y_test_qxt, 'y_pred': y_pred_qxt})
print(results.describe())
```

```python
y_train_qxm = y_train['Qxm']
y_test_qxm = y_test['Qxm']
model_Qxm = LinearRegression()
model_Qxm.fit(X_train, y_train_qxm)
print(model_Qxm.score(X_test, y_test_qxm))

# Dự đoán và đánh giá mô hình
y_pred_qxm = model_Qxm.predict(X_test)
print('MSE:', mean_squared_error(y_test_qxm, y_pred_qxm))
print('RMSE:', np.sqrt(mean_squared_error(y_test_qxm, y_pred_qxm)))
print('MAE:', mean_absolute_error(y_test_qxm, y_pred_qxm))
print(f'MAPE: {mean_absolute_percentage_error(y_test_qxm, y_pred_qxm)*100}%')
results = pd.DataFrame({'y_test': y_test_qxt, 'y_pred': y_pred_qxt})
print(results.describe())
```

```python
y_pred = y_pred_qxm + y_test_qxt
y_test_Tqx = y_test['Tqx']
print('MSE:', mean_squared_error(y_test_Tqx, y_pred))
print('RMSE:', np.sqrt(mean_squared_error(y_test_Tqx, y_pred)))
print('MAE:', mean_absolute_error(y_test_Tqx, y_pred))
print(f'MAPE: {mean_absolute_percentage_error(y_test_Tqx, y_pred)*100}%')
results = pd.DataFrame({'y_test': y_test_Tqx, 'y_pred': y_pred})
print(results.describe())
```
